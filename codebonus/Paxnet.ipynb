## Paxnet Sise Kosdaq

http://www.paxnet.co.kr/stock/sise/jongmok?searchType=kosdaq

```python
import requests
from urllib.parse import urlencode
from bs4 import BeautifulSoup as bts
import pandas as pd
```

```python
url = 'http://www.paxnet.co.kr/stock/sise/siseUpDownAjax'
```

```python
query = {
    'searchType': 'kosdaq',
    'currentPageNo': 1,
    'searchSort': '',
    'searchExchange': 'KOSDAQ',
    'searchTerm': '1',
    'searchUpDown': 'up'
}
```

```python
res = requests.get(url = url, params = query)
```

```python
res.status_code
```

```python
res.headers['content-type']
```

```python
urlencode(query = query)
```

```python
URL = f'{url}?{urlencode(query = query)}'
URL
```

```python
tbl = pd.read_html(io = URL, encoding = 'UTF-8')
```
