```python
import requests
import json
import pandas as pd
```

```python
url = 'https://sports.news.naver.com/kbaseball/news/list'
```

```python
query = {
    'isphoto': 'N', 
    'page': 1, 
    'date': '20230615'
}
```

```python
res = requests.get(url = url, params = query)
```

```python
res.status_code
```

```python
res.headers['content-type']
```

```python
res.text
```

```python
dat = json.loads(s = res.text)
```

```python
dat['list']
```

```python
df = pd.DataFrame(data = dat['list'])
```

```python
url = 'https://section.blog.naver.com/ajax/SearchList.naver'
```

```python
query = {
    'countPerPage': 7,
    'currentPage': 1,
    'endDate': '20230615', 
    'keyword': 'lg evo',
    'orderBy': 'sim',
    'startDate': '20230615',
    'type': 'post'
}
```

```python
headers = {
    'user-agent': 'Mozilla/5.0',
    'referer': 'https://section.blog.naver.com/Search/Post.naver'
}
```

```python
res = requests.get(url = url, params = query, headers = headers)
```

```python
res.status_code
```

```python
res.headers['content-type']
```

```python
res.text.replace(')]}\',\n', '')
```

```python
dat = json.loads(s = res.text.replace(')]}\',\n', ''))
```

```python
dat['result']['searchList']
```

```python
df = pd.DataFrame(data = dat['result']['searchList'])
```
