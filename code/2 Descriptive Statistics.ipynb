### 실습 데이터셋 준비

```python
# 관련 라이브러리를 호출합니다.
import os
import numpy as np
import pandas as pd
```

```python
# 현재 작업경로를 확인합니다.
os.getcwd()
```

```python
# data 폴더로 작업경로를 변경합니다.
os.chdir('../data')
```

```python
# xlsx 파일을 읽고, 데이터프레임 df를 생성합니다.
df = pd.read_excel(io = 'Used_Cars_Price.xlsx')
```

```python
# df의 정보를 확인합니다.
df.info()
```

```python
# df의 일부를 반환합니다.
df.head()
```

### 실습 데이터셋 전처리

```python
# df에서 일부 정수형 변수를 문자형으로 일괄 변환합니다.
cols = ['MetColor', 'Automatic']
df[cols] = df[cols].astype('str')
```

```python
# 열별 자료형을 출력합니다.
df.dtypes
```

### 대푯값

```python
# 연속형 변수의 평균을 반환합니다.
df['Price'].mean().round(2)
```

```python
# 관련 라이브러리를 호출합니다.
from scipy import stats
```

```python
# 양 극단에서 10%씩 제외한 절사평균을 반환합니다.
stats.trim_mean(df['Price'], 0.1).round(2)
```

```python
# 연속형 변수의 중위수를 반환합니다.
df['Price'].median()
```

```python
# [주의] 50% 절사평균을 계산하면 원소 개수에 따라 중위수 또는 결측값을 반환합니다.
stats.trim_mean(df['Price'], 0.5)
```

```python
# [참고] 원소 개수가 짝수이면 결측값을 반환합니다.
stats.trim_mean(range(10), 0.5)
```

```python
# 백분위수를 반환합니다.
df['Price'].quantile(q = np.arange(0, 1.01, 0.01))
```

```python
# 십분위수를 반환합니다.
df['Price'].quantile(q = np.arange(0, 1.01, 0.10))
```

```python
# 사분위수를 반환합니다.
df['Price'].quantile(q = np.arange(0, 1.01, 0.25))
```

```python
# 이산형 또는 범주형 변수의 최빈값을 반환합니다.
df['FuelType'].mode()
```

```python
# 범주형 변수의 빈도수를 내림차순 정렬한 결과를 반환합니다.
df['FuelType'].value_counts()
```

```python
# 빈도수 대신 인덱스명으로 오름차순 정렬한 결과를 반환합니다. 
df['FuelType'].value_counts().sort_index()
```

```python
# 빈도수 대신 백분율을 반환합니다.
df['FuelType'].value_counts(normalize = True)
```

### 산포

```python
# 연속형 변수의 최솟값과 최댓값을 각각 반환합니다.
print(df['Price'].min())
print(df['Price'].max())
```

```python
# 연속형 변수의 범위를 반환합니다.
df['Price'].quantile(q = [0, 1]).diff()
```

```python
# 연속형 변수의 사분범위를 반환합니다.
df['Price'].quantile(q = [0.25, 0.75]).diff()
```

```python
# 연속형 변수의 분산을 반환합니다.
df['Price'].var().round(2)
```

```python
# 연속형 변수의 표준편차를 반환합니다.
df['Price'].std().round(2)
```

```python
# 라이브러리를 호출합니다.
from statsmodels import robust
```

```python
# 연속형 변수의 중위수절대편차를 반환합니다.
robust.mad(df['Price']).round(2)
```

### 여러 열의 기술통계량 확인

```python
# 모든 연속형 변수의 평균을 반환합니다.
df.apply(func = 'mean', axis = 0, numeric_only = True).round(2)
```

```python
# 모든 연속형 변수의 표준편차를 반환합니다.
df.apply(func = 'std', axis = 0, numeric_only = True).round(2)
```

```python
# 모든 연속형 변수의 중위수를 반환합니다.
df.apply(func = 'median', axis = 0, numeric_only = True).round(2)
```

```python
# 모든 연속형 변수의 기술통계량을 반환합니다.
df.describe().round(2)
```

```python
# 모든 범주형 변수의 기술통계량을 반환합니다.
df.describe(include = 'object')
```
