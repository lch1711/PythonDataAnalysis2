```python
import requests
import pandas as pd
```

```python
URL = 'https://bit.ly/KBO_Hitters_2021'
```

```python
res = requests.get(URL)
```

```python
# 디코딩된 문자열
res.text
```

```python
# 인코딩된 문자열
res.content
```

```python
import chardet
```

```python
# 문자열에
chardet.detect(res.content)
```

```python
kbo = pd.read_csv(URL, encoding = 'EUC-KR')
```

```python
kbo.to_csv('KBO_Hitters_2021.csv', index = None)
```
